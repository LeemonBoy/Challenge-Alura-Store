### Importación de datos



```python
# @title
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. **Análisis de facturación**



```python
# @title
facturacion_tienda = tienda.Precio.sum()
facturacion_tienda2 = tienda2.Precio.sum()
facturacion_tienda3 = tienda3.Precio.sum()
facturacion_tienda4 = tienda4.Precio.sum()

print(f"Facturación total de la Tienda 1: {facturacion_tienda}")
print(f"Facturación total de la Tienda 2: {facturacion_tienda2}")
print(f"Facturación total de la Tienda 3: {facturacion_tienda3}")
print(f"Facturación total de la Tienda 4: {facturacion_tienda4}")
```

Gráfico comparativo de facturaciones por tienda

```python
# @title


facturacion = {
                'Tienda 1': facturacion_tienda,
                'Tienda 2': facturacion_tienda2,
                'Tienda 3': facturacion_tienda3,
                'Tienda 4': facturacion_tienda4
              }


plt.figure(figsize=(8,8))
plt.pie(facturacion.values(), labels=facturacion.keys(), autopct='%1.1f%%', startangle=90, colors=['green', 'yellow', 'orange', 'red'])
plt.title('Distribución de Facturación por Tienda')
plt.axis('equal')
plt.show()
```

# 3. **Calificación promedio de la tienda**

# 4. **Productos más y menos vendidos**

# 5. **Envío promedio por tienda**

```python
# @title
costo_promedio_envio_tienda = tienda["Costo de envío"].mean()
costo_promedio_envio_tienda2 = tienda2["Costo de envío"].mean()
costo_promedio_envio_tienda3 = tienda3["Costo de envío"].mean()
costo_promedio_envio_tienda4 = tienda4["Costo de envío"].mean()

print(f"El costo promedio del envio de la Tienda 1 es: {costo_promedio_envio_tienda: .2f}")
print(f"El costo promedio del envio de la Tienda 2 es: {costo_promedio_envio_tienda2: .2f}")
print(f"El costo promedio del envio de la Tienda 3 es: {costo_promedio_envio_tienda3: .2f}")
print(f"El costo promedio del envio de la Tienda 4 es: {costo_promedio_envio_tienda4: .2f}")
```

Gráfico comparativo de promedios de Costos de envio por tienda




```python
# @title
envio_por_tienda = {
                'Tienda 1': costo_promedio_envio_tienda,
                'Tienda 2': costo_promedio_envio_tienda2,
                'Tienda 3': costo_promedio_envio_tienda3,
                'Tienda 4': costo_promedio_envio_tienda4
                   }


plt.figure(figsize=(8,6))
plt.bar(envio_por_tienda.keys(), envio_por_tienda.values(), color='blue')
plt.title('Comparación de promedios de Costos de envio por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Costos de envio')
plt.grid(axis='y')
```
